# Pair Trading Exploration

Interactive companion to the CLI scripts. Run cells top-to-bottom.

Assumes you've already run `python scripts/fetch.py --coins ...` for the coins below.

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent / "src"))

import matplotlib.pyplot as plt
import pandas as pd

from pairtrade.data import load_pair
from pairtrade.analysis import analyze, summarize, print_summary
from pairtrade.screening import screen
from pairtrade.plots import plot_pair, plot_screen

DATA_DIR = Path.cwd().parent / "data"
```

## 1. Analyze one pair

```python
A, B = "bitcoin", "ethereum"
WINDOW, ENTRY, EXIT = 60, 2.0, 0.5

df = load_pair(A, B, DATA_DIR)
result = analyze(df, A, B, window=WINDOW, entry_z=ENTRY, exit_z=EXIT)
print_summary(summarize(result))
result.tail()
```

```python
fig = plot_pair(result, entry_z=ENTRY, exit_z=EXIT)
plt.show()
```

## 2. Screen a basket

Make sure each coin's parquet exists in `data/` (run `scripts/fetch.py` first).

```python
BASKET = ["bitcoin", "ethereum", "solana", "cardano", "ripple", "dogecoin", "litecoin"]
ranked = screen(BASKET, data_dir=DATA_DIR, window=60)
ranked
```

```python
fig = plot_screen(ranked, top=10)
plt.show()
```

## 3. Drill into the best-ranked pair

```python
top = ranked.iloc[0]
A, B = top["a"], top["b"]
result = analyze(load_pair(A, B, DATA_DIR), A, B, window=WINDOW, entry_z=ENTRY, exit_z=EXIT)
print_summary(summarize(result))
plot_pair(result, entry_z=ENTRY, exit_z=EXIT);
```
