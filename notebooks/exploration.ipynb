# Crypto Trend Comparison

Compare two coins' historical price trends.

Assumes you've already run `python scripts/fetch.py --coins ...` for the coins below.

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent / "src"))

import matplotlib.pyplot as plt

from pairtrade.data import load_pair
from pairtrade.analysis import compare, print_summary
from pairtrade.plots import plot_comparison

DATA_DIR = Path.cwd().parent / "data"
```

```python
A, B = "bitcoin", "ethereum"

df = load_pair(A, B, DATA_DIR)
df.tail()
```

## Trend summary

```python
summary = compare(df, A, B)
print_summary(summary)
```

## Side-by-side chart (rebased to 100)

```python
fig = plot_comparison(df, A, B, normalize=True)
plt.show()
```

## Raw prices (dual-axis view)

```python
fig = plot_comparison(df, A, B, normalize=False)
plt.show()
```
